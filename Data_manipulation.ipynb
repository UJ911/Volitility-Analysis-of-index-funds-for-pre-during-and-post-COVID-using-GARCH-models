```python
# prompt: Write python script to 3 seperate time series which captures snp 500 series value before covid, during covid and post covid. use yahoo ticker to access snp 500 data. Plot the 3 different data

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def get_snp_data(start_date, end_date):
    """Downloads SNP 500 data from Yahoo Finance."""
    snp = yf.download("^GSPC", start=start_date, end=end_date)
    return snp

def plot_time_series(data, title):
  plt.figure(figsize=(10, 6))
  plt.plot(data['Adj Close'])
  plt.title(title)
  plt.xlabel('Date')
  plt.ylabel('Adjusted Close Price')
  plt.grid(True)
  plt.show()

if __name__ == "__main__":
    # Define time periods
    pre_covid_start = "2015-01-01"
    pre_covid_end = "2020-02-19"  # Adjust as needed
    covid_start = "2020-02-20"
    covid_end = "2021-10-01"  # Adjust as needed
    post_covid_start = "2021-10-02"
    post_covid_end = "2024-10-31"

    # Download data
    pre_covid_data = get_snp_data(pre_covid_start, pre_covid_end)
    covid_data = get_snp_data(covid_start, covid_end)
    post_covid_data = get_snp_data(post_covid_start, post_covid_end)


    # Plot the data
    plot_time_series(pre_covid_data, "S&P 500 (Pre-Covid)")
    plot_time_series(covid_data, "S&P 500 (During Covid)")
    plot_time_series(post_covid_data, "S&P 500 (Post-Covid)")
```
